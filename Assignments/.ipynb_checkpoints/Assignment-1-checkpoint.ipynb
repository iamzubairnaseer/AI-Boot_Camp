### Q.No.7 Take 5 subject marks, total and generate grade sheet

```python
subjects = ["English", "Computer", "Maths", "Physics", "Biology"]

sub_1 = float(input(f"Enter {subjects[0]} Marks"))
sub_2 = float(input(f"Enter {subjects[1]} Marks"))
sub_3 = float(input(f"Enter {subjects[2]} Marks"))
sub_4 = float(input(f"Enter {subjects[3]} Marks"))
sub_5 = float(input(f"Enter {subjects[4]} Marks"))

obtain_marks = sub_1 + sub_2 + sub_3 + sub_4 + sub_5
total_marks = len(subjects)*100
percentage = (obtain_marks/total_marks)*100

def generate_grade(perc):
    if perc >=80 and perc <= 100:
        grade = "A+"
    elif perc >=70 and perc < 80:
        grade = "A"
    elif perc >=60 and perc < 70:
        grade = "B"
    elif perc >=50 and perc < 60:
        grade = "C"
    elif perc >=40 and perc < 50:
        grade = "D"
    elif perc >=33 and perc < 40:
        grade = "E"
    elif perc >=0 and perc < 33:
        grade = "F"
    else:
        print("Your Percentage is wrong")
        return
    return grade


print(f"""
--------------------------- Mark Sheet ---------------------------

{'Subject':<15}{'Total':>12}{'Obtained':>15}{'Percentage':>15}{'Grade':>10}
-------------------------------------------------------------------

{subjects[0]:<15}{100:>12}{sub_1:>15}{sub_1:>14}%{generate_grade(sub_1):>10}
{subjects[1]:<15}{100:>12}{sub_2:>15}{sub_2:>14}%{generate_grade(sub_2):>10}
{subjects[2]:<15}{100:>12}{sub_3:>15}{sub_3:>14}%{generate_grade(sub_3):>10}
{subjects[3]:<15}{100:>12}{sub_4:>15}{sub_4:>14}%{generate_grade(sub_4):>10}
{subjects[4]:<15}{100:>12}{sub_5:>15}{sub_5:>14}%{generate_grade(sub_5):>10}
--------------------------------------------------------------
Total Marks:          {total_marks}
Total Obtained Marks: {obtain_marks}
Overall Percentage:   {percentage:.2f}%
Overall Grade:        {generate_grade(int(percentage))}
""")
```

### Q.No.8 Check if number is even or odd

```python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(f"The number({number}) is Even")
else:
    print(f"The number({number}) is Odd")
```
