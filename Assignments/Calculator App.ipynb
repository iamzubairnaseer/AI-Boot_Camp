```python
print("-----------------CALCULATOR-----------------")
num1 = float(int(input("Enter the first number: ")))
num2 = float(int(input("Enter the second number: ")))
print("ALLOWED OPERATION: + - x /")
opt = input("Enter the operator: ")

if opt == "+":
    answer = num1+num2
elif opt == "-":
    answer = num1-num2
elif opt == "*" or opt == "x" or opt == "X":
    answer = num1*num2
elif opt == "/":
    if num2==0:
        answer = "denomitor can't be 0"
    else:
        answer = num1/num2
else:
    answer = "Wrong operation"

print(f"{num1} {opt} {num2} = {answer}")
```
